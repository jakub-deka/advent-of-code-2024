```python
from collections import namedtuple

Coord = namedtuple("Coord", ["x", "y"])


def read_file(path):
    res = open(path, "r").readlines()
    res = [Coord(*tuple(map(int, x.strip().split(",")))) for x in res]
    # swap to get the same as example
    res = [Coord(x.y, x.x) for x in res]
    return res


def populate_grid(path, height, width, block_count):
    grid = [["." for _ in range(width)] for _ in range(height)]
    blocks = read_file(path)
    for b in blocks[:block_count]:
        grid[b.x][b.y] = "#"

    return grid
```

```python
from collections import deque

# grid = populate_grid("example", 7, 7, 12)
grid = populate_grid("input", 71, 71, 1024)
START = (0, 0)
END = (70, 70)
DIRECTIONS = [(0, 1), (1, 0), (0, -1), (-1, 0)]


def bfs(grid):
    queue = deque()
    queue.append((*START, 0))
    visited = set()

    while queue:
        x, y, length = queue.popleft()
        if (x, y) in visited:
            continue

        visited.add((x, y))
        if (x, y) == END:
            return length
        for dx, dy in DIRECTIONS:
            nx, ny = x + dx, y + dy
            if 0 <= nx < len(grid) and 0 <= ny < len(grid[0]) and grid[nx][ny] == ".":
                queue.append((nx, ny, length + 1))


bfs(grid)
```
